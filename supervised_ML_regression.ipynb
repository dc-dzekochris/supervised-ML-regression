```python
#sales prediction based on advertising
#importing pandas libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading dataset
sales_data=pd.read_csv('https://msi.martial.co.ke/data/Advertising.csv')
sales_data.head()
```

```python
#data analysis
sales_data.info()
```

```python
#correlation analysis
sales_data.corr()
```

```python
#machine learning
#step 1: divide dataset into features(x) and label(y)
#convert sales dataset into numpy array to extract values
array=sales_data.values
array
#x=features(tv,radio,newspaper)
#200 records and 3 columns
x=array[:,0:3]
x
print(x.shape)

y=array[:,3]
print(y.shape)
```

```python
#step 2
#divide the x and y into training and testing
#x->x_train(70) and x_test(30)
#y->y-train(70) and y_test(30)

from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.30,random_state=42)
print (x_train.shape,y_train.shape)
print (x_test.shape,y_test.shape)
```

```python
#step 3
#import machine learning algorithm
#a. linear regression
from sklearn.linear_model import LinearRegression

#b. decision trees
from sklearn.tree import DecisionTreeRegressor

#c. k-nearest neighbor algorithm
from sklearn.neighbors import KNeighborsRegressor

#d. support vector machine
from sklearn.svm import SVR

#e. random forest
from sklearn.ensemble import RandomForestRegressor

#f. gradient boosting
from sklearn.ensemble import GradientBoostingRegressor

#g. adaboost
from sklearn.ensemble import AdaBoostRegressor

```

```python
#step 4: cross validating
#process that is used to determine the best learning algorithm for the particular problem
# STEP4: CROSS VALIDATION
# Hyperparameter Tuning.
# Process that is used to Determine the Best Learning Algorithm for a Particular Problem.
models= []
models.append(('D Trees', DecisionTreeRegressor()))
models.append(('KNR', KNeighborsRegressor()))
models.append(('LR'models, LinearRegression()))
models.append(('Support Machines', SVR()))
models.append(('Random Forest', RandomForestRegressor()))
models.append(('Gradient Boosting', GradientBoostingRegressor()))
models.append(('Support Machines', SVR()))

# Import Cross Validation and KFOLD
from sklearn.model_selection import cross_val_score, KFold
# Create a for loop so that each model is tested in turn
for name, model in models:
          # Here we do a 10 split K-FOLD
          kfold = KFold(n_splits = 10, random_state=42, shuffle=True)

          # We get the results for each Fold, Train and Test using the Folds
          cv_results = cross_val_score(model, x_train, y_train, cv = kfold,
          scoring='r2')

          # Get the average of all Folds for current model in the loop
          print(name, ' Results:= ', cv_results.mean())
```

```python
#step 5: model training
#in this process, the models learnt from the training dataset using the best algorithm
model= GradientBoostingRegressor()
model.fit(x_train,y_train)
```

```python
#step 6: model evaluation
# process used to determine whether model is accurate to predict the results
#in regression(numbers), we use the Mean Absolute Error(mae)
#to evaluate a model we will provide the testing features(x_test)and have to hide the testing outcome(y_test)
#model will provide its own predictions(model_predictions), that will be compared against the correct predictions(Y_test)

model_predictions = model.predict(x_test)
print("The Model Predicted: ", model_predictions)

print("\n\n")

print("The Correct Answer is : ", y_test)
```

```python
# Therefore, we will a scientific called r2_score, which is the ratio between the model_prediction and Y_test result
from sklearn.metrics import r2_score
accuracy = r2_score(y_test, model_predictions) * 100
print("The Accuracy of the Model is: ", accuracy)
```
